Se elimina las filas donde no hay datos de las columnas Name, Year_of_Release, Publisher y Genre, ya que no superan el 2% del total de los datos.

```python
#Creamos una compia de la base original
df_limpio = df
```

###**Cálculo de filas y columnas después de la limpieza**

```python
num_filas, num_columnas = df_limpio.shape

print(f"Número de filas: {num_filas}")
print(f"Número de columnas: {num_columnas}")
```

La información sobre el df original es:

```python
# Convierte la columna 'User_Score' a tipo numérico, estableciendo como NaN los valores no convertibles
df_limpio['User_Score'] = pd.to_numeric(df_limpio['User_Score'], errors='coerce')

# Calcula la media de las variables
mean_Critic_Score = df_limpio['Critic_Score'].mean()
mean_Critic_Count = df_limpio['Critic_Count'].mean()
mean_User_Score = df_limpio['User_Score'].mean()
mean_User_Count = df_limpio['User_Count'].mean()

# Reemplaza los valores NaN con las medias calculadas
df_limpio['Critic_Score'].fillna(mean_Critic_Score, inplace=True)
df_limpio['Critic_Count'].fillna(mean_Critic_Count, inplace=True)
df_limpio['User_Score'].fillna(mean_User_Score, inplace=True)
df_limpio['User_Count'].fillna(mean_User_Count, inplace=True)

missing_values = df_limpio.isnull().sum()
print(missing_values)
```

## **Gráfico de barras para los 10 juegos más vendidos a nivel global**

```python
# Creación del gráfico de barras para los 10 juegos más vendidos a nivel global

import seaborn as sns
import matplotlib.pyplot as plt

juegos_mas_vendidos = df_limpio.sort_values(by='Global_Sales', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

sns.barplot(x='Global_Sales', y='Name', data=juegos_mas_vendidos, palette="viridis")

plt.title('Los 10 juegos más vendidos globalmente')
plt.xlabel('Ventas globales (en millones de USD)')
plt.ylabel('Nombre del Juego')

plt.show()
```

### **Insight**
Si queremos hacer un top 3 de los juegos más vendidos, podríamos decir que Wii Sports es el más vendido, seguido muy por debajo de "Super Mario Bros" y "Mario Kart Wii". El primero con más de 80 millones de dólares en ventas, el segundo con alrededor de 40 millones y el último con apróximadamente 36 millones de dólares.

## **Gráfico de anillo Distribución de Ventas por Género de video juego**

```python
ventas_genero = df.groupby('Genre')['Global_Sales'].sum().reset_index()

custom_palette = sns.color_palette("husl", len(ventas_genero))

plt.figure(figsize=(10, 10))
plt.pie(
    ventas_genero['Global_Sales'],
    labels=ventas_genero['Genre'],
    autopct='%1.1f%%',
    startangle=140,
    colors=custom_palette,
    textprops={'fontsize': 12}
)

plt.title('Distribución de ventas por Género', fontsize=16)

plt.gca().add_artist(plt.Circle((0, 0), 0.70, fc='white'))

plt.axis('equal')
plt.show()
```

###**Insight**
Se observa que el genero más vendido es Acción en un 19,5%, seguido de Deportes en 14,9% y el genero Shooter con un 11,8 %.

## **Gráfico 10 Plataformas más utilizadas**

## **Gráfico Ventas globales por género a lo largo del tiempo**

```python
import pandas as pd
import matplotlib.pyplot as plt

genre_sales = df.groupby(['Year_of_Release', 'Genre'])['Global_Sales'].sum().unstack()
genre_sales.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Ventas Globales por Género a lo largo de los Años')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Ventas Globales')
plt.legend(title='Género', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

###**Insight**
La distribución de ventas globales tiene un pico en el año 2008, lo que muestra que en ese año se lazaron más videojuegos con más ventas globales.

Como observamos en la gráfica anterior vemos que el género "sports" ha sido uno de los más vendido a lo largo de los años.

Realizando un analisis de las evolución de mercado vemos que las ventas de videojueos han disminuido desde el 2012 hasta el año 2020, a causa de los cambios en la tecnología y la industria del entretenimiento.

# Popularidad de los desarrolladores

```python
#cantidad de juegos por desarrollador
developer_counts = df_limpio['Developer'].value_counts().reset_index()
developer_counts.columns = ['Developer', 'Count']

# Calcular las ventas totales por desarrollador
developer_sales = df_limpio.groupby('Developer')['Global_Sales'].sum().reset_index()

# Unir ambos DataFrames
merged_data = developer_counts.merge(developer_sales, on='Developer')

# Filtrar para los 10 desarrolladores más populares
top_developers = merged_data.nlargest(10, 'Count')

# Crear un gráfico
plt.figure(figsize=(12, 8))
plt.scatter(top_developers['Count'], top_developers['Global_Sales'], s=top_developers['Global_Sales']*10, alpha=0.5)


for i, row in top_developers.iterrows():
    plt.text(row['Count'], row['Global_Sales'], row['Developer'], fontsize=10, ha='right')

plt.title('Popularidad de los Desarrolladores (Top 10)')
plt.xlabel('Número de Juegos Lanzados')
plt.ylabel('Ventas Globales')
plt.show()
```

**Insight**

Se observa que EA Sport es el desarrollor que más ventas ha tenido con menor numero de juegos lanzados.
